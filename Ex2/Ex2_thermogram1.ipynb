```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def invertChannels(img):
    b,g,r = cv2.split(img) # split channels
    return cv2.merge([r,g,b]) # merge in rgb order to display with matplotlib
```

```python
# mean square error
def mse(imageA, imageB):
    # the 'Mean Squared Error' between the two images is the
    # sum of the squared difference between the two images;
    # NOTE: the two images must have the same dimension
    err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
    err /= float(imageA.shape[0] * imageA.shape[1])

    # return the MSE, the lower the error, the more "similar"
    return err
```

```python
img_original = cv2.imread('../colorImages/face_thermogram.jpg')
img = cv2.imread('../colorImages/face_thermogram_1.jpg')

plt.figure(figsize=(10,5)) 

plt.subplot(121)
plt.imshow(invertChannels(img_original))
plt.title('Original Image')

plt.subplot(122)
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img_original],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])   
plt.title('Original Image')   
```

```python

plt.figure(figsize=(10,5)) 
plt.subplot(121)
plt.imshow(invertChannels(img))
plt.title('Thermogram 1')

plt.subplot(122)
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])    
plt.title('Thermogram 1') 
```

# Imagem invertida

```python
plt.figure(figsize=(5,5)) 
plt.imshow(invertChannels(255 - img))
plt.title('Enhanced Image')
```

```python
plt.figure(figsize=(5,5)) 
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([255 - img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])   
plt.title('Enhanced Image')
```

```python
# mean square error comparation
print "MSE between Original and Thermogram1: %s" % mse(img_original,img)
print "MSE between Original and Thermogram1 enchanced: %s" % mse(img_original,255 - img)
```
